#　经验分布函数

## 经验分布函数的性质

### 经验分布函数的矩

对于每个固定的 $x$，有

$$
    \mathbb{E} \left[ \hat F_n(x) \right] = F(x), \quad \text{Var}\left( \hat F_n(x) \right) = \frac{F(x)(1-F(x))}{n}
$$

并因此 $ \hat F_n(x) \xrightarrow{P} F(x)$。

#### 证明过程

$$
    \begin{aligned}
        \mathbb{E} \left[ \hat F_n(x) \right] & = \mathbb{E} \left[ \frac{1}{n} \sum_{i=1}^n I(X_i \leq x) \right]\\
        & = \frac{1}{n} \sum_{i=1}^n \mathbb{E} \left[ I(X_i \leq x) \right] \\
        & = \frac{1}{n} \sum_{i=1}^n F(x) \\
        & = F(x),
    \end{aligned}
$$

$$
    \begin{aligned}
        \text{Var} \left( \hat{F}_n(x) \right) & = \text{Var} \left( \frac{1}{n} \sum_{i=1}^n I(X_i \leq x) \right)  \\
        & = \frac{1}{n^2} \sum_{i=1}^n \text{Var} \left( I(X_i \leq x) \right) \\
        & = \frac{1}{n^2} \sum_{i=1}^n \left( F(x)(1-F(x)) \right) \\
        & = \frac{F(x)(1-F(x))}{n},  
    \end{aligned}
$$

由弱大数定律可知，

$$
    \hat{F}_n(x) = \frac{1}{n} \sum_{i=1}^n I(X_i \leq x) \xrightarrow{P} \mathbb{E} \left[ I(X \leq x) \right] = F(x),
$$

因此，结论得证。

### Glivenko–Cantelli 定理

对于一个连续随机变量 $X$，记其 $\text{CDF}$ 为 $F(x)$，则基于 $n$ 个独立同分布样本的经验分布函数 $\hat{F}_n(x)$ 满足

$$
    \sup_{x} | \hat{F}_n(x) - F(x) | \xrightarrow{a.s.} 0, \quad n \to \infty.
$$

#### 证明过程

考虑一个连续随机变量 $X$（实际上若分布中存在离散型分布，证明依然成立），我们按照 $\text{cdf}$ 将其划分为 $m$ 个区间，即定义 $(-\infty = x_0 < x_1 < \cdots < x_m < x_{m+1} = \infty)$，满足

$$F(x_j) - F(x_{j-1}) = \frac{1}{m}, \quad j = 1, \ldots, m.$$

现在 $\forall x \in \mathbb{R}$，存在 $j \in {1, \ldots, m}$ 使得 $x \in [x_{j-1}, x_j)$。

那么对于 $x \in [x_{j-1}, x_j)$，有：

$$\hat{F}_n(x) - F(x) \leq \hat{F}_n(x_j) - F(x_{j-1}) = \hat{F}_n(x_j) - F(x_j) + \frac{1}{m},$$

$$\hat{F}_n(x) - F(x) \geq \hat{F}_n(x_{j-1}) - F(x_j) = \hat{F}_n(x_{j-1}) - F(x_{j-1}) - \frac{1}{m}.$$

因此，

$$\sup_{x} | \hat{F}_n(x) - F(x) | \leq \max_{j \in \{1, \ldots, m\}} | \hat{F}_n(x_j) - F(x_j) | + \frac{1}{m}.$$

由强大数定律知，$\max_{j \in {1, \ldots, m}} | \hat{F}n(x_j) - F(x_j) | \xrightarrow{a.s.} 0$。对于 $\forall \epsilon > 0$，当 $m$ 足够大时，可以找到 $N$ 使得当 $n > N$ 时，

$$\max_{j \in {1, \ldots, m}} | \hat{F}_n(x_j) - F(x_j) | < \epsilon - \frac{1}{m} \quad a.s..$$

结合上述结果，这意味着

$$ \sup_{x} | \hat{F}_n(x) - F(x) | < \epsilon - \frac{1}{m} + \frac{1}{m} = \epsilon \quad \forall n > N,$$

由于 $m$ 可以取任意大值，$\sup_{x} | \hat{F}_n(x) - F(x) | \xrightarrow{a.s.} 0$ 。

因此，结论得证。

实际上，我们想要的结论的收敛性是几乎处处收敛，因此自然联想到了强大数定律。我们通过划分区间，利用强大数定律找到了经验分布函数和累积分布函数的差的上界。我们用强大数定律简化对 $\hat F_n(x,\omega)$ 的讨论为 $\hat F_n(x)$，其中 $\omega \in \Omega$，$\Omega$ 为事件空间。

这个定理给了我们一个更好的收敛性的刻画，这是一种一致收敛，在 $n \rightarrow \infty$ 的时候，我们期望的是所有的点都按相同的速度收敛到 $0$，下文的 DKW 不等式给了这个速度一个更明确的刻画。

### Dvoretzky-Kiefer-Wolfowitz (DKW) 不等式

对于任何的 $\epsilon > 0$，

$$
    P \left( \sup_x \left| \hat F_n(x) - F(x) \right| > \epsilon \right) \leq 2e^{-2n\epsilon^2}.
$$

这个定理的证明较为复杂，具体过程可以参见 [Dvoretzky, A.; Kiefer, J.; Wolfowitz, J. (1956)](https://projecteuclid.org/journals/annals-of-mathematical-statistics/volume-27/issue-3/Asymptotic-Minimax-Character-of-the-Sample-Distribution-Function-and-of/10.1214/aoms/1177728174.full) 和 [Massart, P. (1990)](https://projecteuclid.org/journals/annals-of-probability/volume-18/issue-3/The-Tight-Constant-in-the-Dvoretzky-Kiefer-Wolfowitz-Inequality/10.1214/aop/1176990746.full) 。

笔者这边提供一个思路，不一定严谨。注意到对于每一个 $ x $，考虑 $ X_i $ 是服从分布 $F$ 的随机变量，那么有 $ I(X_i \leq x)$ 服从 Bernoulli 分布， 即 $ I(X_i \leq x) \sim \text{Bernoulli}(F(x))$。那么根据 Hoeffding 不等式可以得到 

$$
    P \left( \left| \hat F_n(x) - F(x) \right| > \epsilon \right) = P \left( \left| \frac{1}{n} \sum_{i=1}^n I(X_i \leq x) - F(x) \right| > \epsilon \right) \leq 2e^{-2n\epsilon^2}, \forall x.
$$

由于对于任意的 $ x $，这个不等式均成立，且这个上界并没有依赖 $x$ 的取值，因此如果对 $ x $ 取上确界，这个不等式仍然成立。

即

$$
    P \left( \sup_x \left| \hat F_n(x) - F(x) \right| > \epsilon \right) \leq 2e^{-2n\epsilon^2}.
$$

DKW 不等式为我们提供了一个一致收敛速度的保证，同时也是从数值上说明了 Glivenko–Cantelli 定理的正确性。

## 经验分布函数的置信集

### 置信集

若记 $ \mathfrak{F} $ 为分布函数 $ F $ 的一个类，而 $ \theta $ 为某个感兴趣的量。这样，$ \theta $ 可能是 $ F $ 本身，或者是 $ F' $，或者 $ F $ 的均值等。令 $ C_n$ 为 $ \theta $ 可能取值的集合，它依赖于数据 $ X_1, \cdots, X_n $。为了强调概率的陈述依赖于背景中的 $ F $，有时记它为 $ P_F $。

如果

$$ \inf_{F \in \mathfrak{F}} P_F(\theta \in C_n) \geq 1 - \alpha, \quad \forall n \in \mathbb{N} $$

那么 $ C_n $ 为一个有穷样本 $ 1 - \alpha $ 置信集。

如果

$$ \liminf_{n \to \infty} \inf_{F \in \mathfrak{F}} P_F(\theta \in C_n) \geq 1 - \alpha, $$

那么 $ C_n $ 为一个一致渐进 $ 1 - \alpha $ 置信集。

如果

$$ \liminf_{n \to \infty} P_F(\theta \in C_n) \geq 1 - \alpha, \quad \forall F \in \mathfrak{F} $$

那么 $ C_n $ 为一个逐点渐进 $ 1 - \alpha $ 置信集。

这里我们区分这三种置信集的原因是，这三种置信集的严格程度递减，我们在构造置信集的时候需要明确这个置信集的效力。对于有穷样本置信集，不依赖于样本量 $ n \to \infty $ 的渐进性质，而是基于有限样本的严格概率界，因此置信带相对最宽。对于一致渐进置信集，覆盖概率在整个 $ x \in \mathbb{R} $ 上均匀收敛，但是需要 $ n \to \infty $。对于逐点渐进置信集，只保证在每个固定 $ x $ 处，置信区间覆盖 $ F(x) $ 的概率趋向于 $ 1-\alpha $，但无法保证整个函数的联合覆盖概率，因为不同 $ x $ 之间的 $ F_n(x) $ 是相关的。

### DKW 置信集

由 DKW 不等式，能够构造一个置信集。令 $ \epsilon_n = \sqrt{\frac{1}{2n} \ln \frac{2}{\alpha}} $，$ L(x) = \min \{ \hat F_n(x) - \epsilon_n, 0\} $，$ U(x) = \max \{ \hat F_n(x) - \epsilon_n, 1\} $。则 $ (L(x), U(x)) $ 是一个非参数 $ 1 - \alpha $ 置信带。


下面是根据 DKW 不等式构造的置信集。

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 设置参数
n = 1000  # 样本量
alpha = 0.05  # 显著性水平
num_points = 1000  # 用于绘制曲线的点数

# 生成标准正态分布的随机样本
np.random.seed(0)  # 为了可重现性
data = np.random.normal(0, 1, n)

# 计算经验分布函数
x = np.linspace(-3, 3, num_points)  # x轴范围
F_n = np.array([np.mean(data <= xi) for xi in x])  # 经验分布函数

# 根据DKW不等式计算epsilon
epsilon_n = np.sqrt((1/(2*n)) * np.log(2/alpha))

# 计算置信带上下界
L_x = np.maximum(F_n - epsilon_n, 0)  # 下界
U_x = np.minimum(F_n + epsilon_n, 1)  # 上界

# 真实标准正态分布的CDF
true_cdf = norm.cdf(x)

# 绘图
plt.figure(figsize=(10, 6))
plt.plot(x, F_n, label='Empirical CDF', color='blue')
plt.plot(x, L_x, 'r--', label=f'{1-alpha:.2f} Confidence Band (Lower)')
plt.plot(x, U_x, 'r--', label=f'{1-alpha:.2f} Confidence Band (Upper)')
plt.plot(x, true_cdf, 'k-', label='True CDF (N(0,1))')
plt.fill_between(x, L_x, U_x, color='red', alpha=0.1)
plt.title(f'DKW Confidence Band for Standard Normal (n={n}, α={alpha})')
plt.xlabel('x')
plt.ylabel('CDF')
plt.legend()
plt.grid(True)
plt.show()
```

可以看到，这个置信带较宽，并且成功覆盖了我们得到的经验分布函数。

下面是根据 $\hat F_n(x)$ 估计 $ \text{SE} = \sqrt{\frac{\hat F_n(x) (1 - \hat F_n(x))}{n}}$，再利用大样本的中心极限定理构造的置信带。

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 设置参数
n = 1000  # 样本量
alpha = 0.05  # 显著性水平
num_points = 1000  # 用于绘制曲线的点数

# 生成标准正态分布的随机样本
np.random.seed(0)  # 为了可重现性
data = np.random.normal(0, 1, n)

# 计算经验分布函数
x = np.linspace(-3, 3, num_points)  # x轴范围
F_n = np.array([np.mean(data <= xi) for xi in x])  # 经验分布函数

# 计算标准误
se = np.sqrt(F_n * (1 - F_n) / n)  # se = sqrt(F_n(x) * (1 - F_n(x)) / n)

# 计算逐点置信带
z = norm.ppf(1 - alpha / 2)  # 正态分布临界值
L_x_pointwise = np.maximum(F_n - z * se, 0)  # 逐点下界
U_x_pointwise = np.minimum(F_n + z * se, 1)  # 逐点上界

# 真实标准正态分布的CDF
true_cdf = norm.cdf(x)

# 绘图
plt.figure(figsize=(10, 6))
plt.plot(x, F_n, label='Empirical CDF', color='blue')
plt.plot(x, L_x_pointwise, 'r--', label=f'{1-alpha:.2f} Pointwise Confidence Band (Lower)')
plt.plot(x, U_x_pointwise, 'r--', label=f'{1-alpha:.2f} Pointwise Confidence Band (Upper)')
plt.plot(x, true_cdf, 'k-', label='True CDF (N(0,1))')
plt.fill_between(x, L_x_pointwise, U_x_pointwise, color='red', alpha=0.1)
plt.title(f'Pointwise Confidence Band for Standard Normal (n={n}, α={alpha})')
plt.xlabel('x')
plt.ylabel('CDF')
plt.legend()
plt.grid(True)
plt.show()
```

可以看到，置信带较窄，在有些地方真实分布会超过置信带，因此这个逐点置信带并不是一个好的置信带。

这本质上是因为，我们对于每个 $x$ 都计算了一个置信区间，但此时所有的 $ x $ 都落在区间内的概率并不是 $ 1 - \alpha $，这个现象有些类似于生物统计中对 $ p $ 值做一些修正的背后的原因，都是多次实验导致联合概率并不是原来的 $ 1 - \alpha $。因此，我们需要使用 DKW 不等式来构造有穷样本置信带。
